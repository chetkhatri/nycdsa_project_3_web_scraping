```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from sklearn.cluster import KMeans
```

```python
park_data_frame = pd.read_csv('./park_data_frame.csv')
park_data_frame_name = park_data_frame["name"]
park_data_frame = park_data_frame.drop(["name"],axis=1)
```

```python
park_data_frame_2 = pd.get_dummies(park_data_frame)
# park_data_frame_2.keys()
km = KMeans(n_clusters=5,
            init='random',
            n_init=10,
            max_iter=300,
            tol=0.0001,
            random_state=0)
park_data_frame_class = km.fit_predict(park_data_frame_2)
park_data_frame_2["class"] = park_data_frame_class
park_data_frame_2["name"] = park_data_frame_name 
# park_data_frame_2.keys()
park_data_frame_final = park_data_frame
park_data_frame_final[["name","class"]] = park_data_frame_2[["name","class"]]
park_data_frame_final = park_data_frame_final.reindex_axis(sorted(park_data_frame_final.columns), axis=1)
park_data_frame_final
```

```python
park_data_frame_wiki = pd.read_csv('./park_data_frame_wiki.csv')
#park_data_frame_wiki
```

```python
## groupby 
grouped = park_data_frame_wiki.groupby(['state'])
bins = pd.DataFrame({'state': grouped['name'].count(),'visitor': grouped['visitor'].mean()})
new_frame = pd.DataFrame({'state': bins.index.values,'count':grouped['name'].count().values,'Visitor': grouped['visitor'].mean().values})
new_frame.to_csv('./df_state_count.csv', index=False)
#new_frame
```
